```python
from sklearn.model_selection import train_test_split
import glob
import os
from shutil import copyfile

dataset_labels_path = '/media/aivn24/partition1/HungAn/Tuong/OD/Dataset_Bosch_OD/all'

image_list = glob.glob(f'{dataset_labels_path}/*.txt')
train_list, val_list = train_test_split(image_list, test_size=0.2, random_state=42)

print('total = ', len(image_list))
print('train :', len(train_list))
print('val :', len(val_list))
```

```python
path_train = '/media/aivn24/partition1/HungAn/Tuong/OD/Dataset_Bosch_OD/train'
path_val = '/media/aivn24/partition1/HungAn/Tuong/OD/Dataset_Bosch_OD/val'
os.makedirs(f'{path_train}/images', exist_ok=True)
os.makedirs(f'{path_train}/labels', exist_ok=True)
os.makedirs(f'{path_val}/images', exist_ok=True)
os.makedirs(f'{path_val}/labels', exist_ok=True)

for filename in train_list:
    image_name = filename.split('/')[-1]
    image_name = image_name.split('.')
    image_name = '.'.join(image_name[:-1])
    copyfile(os.path.join(dataset_labels_path, image_name +'.png'), os.path.join(f'{path_train}/images', image_name +'.png'))
    copyfile(filename, os.path.join(f'{path_train}/labels', image_name+'.txt'))

for filename in val_list:
    image_name = filename.split('/')[-1]
    image_name = image_name.split('.')
    image_name = '.'.join(image_name[:-1])
    copyfile(os.path.join(dataset_labels_path, image_name +'.png'), os.path.join(f'{path_val}/images', image_name +'.png'))
    copyfile(filename, os.path.join(f'{path_val}/labels', image_name+'.txt'))
```
